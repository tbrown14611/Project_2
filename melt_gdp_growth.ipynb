```python
# Import libraries
import pandas as pd
```

```python
# Load  the CSV file into a DataFrames
# Import the dataset
gdp_growth_df = pd.read_csv('Working Files/gdp_growth.csv')
```

```python
display(gdp_growth_df.head())
```

```python
# Reset the index so 'Country' is a column
gdp_growth_df_reindexed = gdp_growth_df.reset_index()
gdp_growth_df_reindexed
```

```python
# Convert the DataFrame from short form to long form
# Melt the DataFrame
gdp_growth_df_reindexed.melt()
```

```python
# Melt the DataFrame and rename the columnns
melted_growth_df = gdp_growth_df_reindexed.melt(id_vars=['Country Name',
                                                           'Country Code'],
                                                  var_name='Year',
                                                  value_name='GDP Growth')

melted_growth_df
```

```python
# Save the updated dataframe to a CSV file
updated_melted_growth_df_path = 'Working Files/updated_growth_gdp.csv'
melted_growth_df.to_csv(updated_melted_growth_df_path, index=False)


```
