```python
# Set start time

from datetime import datetime
start_time = datetime.now()
```

```python
# Import libraries
import pandas as pd
```

```python
# Load the CSV file
resort1_df = pd.read_csv('Resources/result_files/step1_build_exchange_gdp_sorted_df.csv')
```

```python
# Display the first five rows of the DataFrame
resort1_df.head()
```

```python
# Melt the dataframe to convert year columns into rows and align them with the 'Year' column
df_melted = resort1_df.melt(id_vars=['Record Date', 'Exchange Rate', 'Effective Date', 'Country', 'Currency', 'Country Name', 'Country Code', 'Year'],
                    value_vars=[str(year) for year in range(2001, 2024)],
                    var_name='Year_temp', value_name='GDPGrowth')
```

```python
# Ensure the 'Year_temp' is numeric and matches the 'Year' column
df_melted['Year_temp'] = df_melted['Year_temp'].astype(int)
```

```python
# Filter rows where 'Year' matches 'Year_temp'
df_filtered = df_melted[df_melted['Year'] == df_melted['Year_temp']]
```

```python
# Drop the 'Year_temp' column
df_filtered = df_filtered.drop(columns=['Year_temp'])
```

```python
# Reorder columns
df_final = df_filtered[['Record Date', 'Exchange Rate', 'Effective Date', 'Country', 'Currency', 'Country Name', 'Country Code', 'GDPGrowth', 'Year']]
```

```python
# Save the resulting dataframe to a new CSV file
output_path = 'Resources/result_files/step2_reformat_data_for_analysis.csv'
df_final.to_csv(output_path, index=False)
```
